```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from config import g_key

gmaps.configure(api_key=g_key)

file_to_load = "~/python-api-challenge/WeatherPy/output_data_cities.csv"
```

```python
weather_data_df = pd.read_csv(file_to_load) 
weather_data_df.head()
```

```python
locations = weather_data_df[["Lat", "Lng"]]
hum_level = weather_data_df["Humidity"].astype(float)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=hum_level, dissipating=False, max_intensity=100,point_radius=1)

fig.add_layer(heat_layer)

fig
```

```python
n_weather_data_df = weather_data_df.loc[(weather_data_df["Max Temp (F)"]> 70)&(weather_data_df["Max Temp (F)"]< 80)&(weather_data_df["Wind Speed"]< 5)&(weather_data_df["Cloudiness"]< 25)]
n_weather_data_df.head()
```

```python
n_weather_data_df["Hotel Name"] = ""
```

```python
hotel_df = n_weather_data_df
hotel_df
```

```python
params = {
    "radius": 5000,
    "types": "hotel",
    "Keywords": "hotels",
    "key": g_key
}

for index, row in hotel_df.iterrows():

    lat = row["Lat"]
    lng = row["Lng"]

    params["location"] = f"{lat},{lng}"

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    name_address = requests.get(base_url, params=params)

    name_address = name_address.json()

    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["results"][0]["name"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```

```python
hotel_df
```

```python
hotel_name = hotel_df["Hotel Name"].tolist()
city = hotel_df["City Name"].tolist()
country = hotel_df["Country"].tolist()
```

```python
marker_locations = hotel_df[['Lat', 'Lng']]

info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""

fig = gmaps.figure()
markers = gmaps.marker_layer(marker_locations,info_box_content =info_box_template)
fig.add_layer(markers)
fig
```
